```python
import pandas as pd
import matplotlib.pyplot as plt

def drawing_pie(local,male_count,female_count):
    plt.rc('font', family='Malgun Gothic')
    color = ['cornflowerblue','tomato']
    population = male_count + female_count
    plt.pie(population, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
    plt.title(f'{local} 남녀 성별 비율')
    plt.show()
```

```python
df = pd.read_csv('gender.csv')
df = df[df['행정구역'].str[:2]=='대구']
df = df.iloc[:,[0,14,27]]
df['행정구역']=df['행정구역'].str[:-12]
df = df.set_index('행정구역')
```

```python
pie_0 = df[df.index=='대구광역시  ']
pie_1 = df[df.index=='대구광역시 중구 ']
pie_2 = df[df.index=='대구광역시 동구 ']
pie_3 = df[df.index=='대구광역시 서구 ']
pie_4 = df[df.index=='대구광역시 남구 ']
pie_5 = df[df.index=='대구광역시 북구 ']
pie_6 = df[df.index=='대구광역시 수성구 ']
pie_7 = df[df.index=='대구광역시 달서구 ']
pie_8 = df[df.index=='대구광역시 달성군 ']
lists=[pie_0,pie_1,pie_2,pie_3,pie_4,pie_5,pie_6,pie_7,pie_8]
```

```python
for x in range(len(lists)):
    lists[x] = lists[x].iloc[:,:].applymap(lambda x: int(x.replace(',','')))
```

```python
ax = plt.figure(figsize=(15,15))
plt.rc('font', family='Malgun Gothic')
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.pie([lists[i].iloc[:,0][0],lists[i].iloc[:,1][0]],autopct='%.1f%%',labels=['남성','여성'], startangle=90)
    plt.title(lists[i].index[0])
ax.suptitle("대구광역시 구별 남녀 인구 비율",fontsize=20)
plt.show()
for i in range(9):
    print(f'{lists[i].index[0]}: [{lists[i].iloc[:,0][0]},{lists[i].iloc[:,1][0]}]')
```
