```python
from konlpy.tag import Okt

okt = Okt()
```

```python
splited = []
for i in df['data']:
    splited.append(i.split(' '))
```

```python
df['split']=splited
```

```python
df
```

```python
from tqdm import tqdm
```

```python
from konlpy.tag import Okt

okt = Okt()

word_list = []

for line in tqdm(df['data']):
    words = []
    for w,t in okt.pos(line):
        if t in ['Verb','Adjective']:
            words.append(w)
        elif (t == 'Punctuation') and ('*' in w):
            words.append(w)
        
    sent = ' '.join(words)
    word_list.append(sent)
```

```python
words=[]
for i in tqdm(range(len(word_list))):
    words+=(word_list[i].split(' '))
```

```python
words
```

```python
dic = {}

for i in words:
    if i not in dic:
        dic[i] = 1
    else:
        dic[i] +=1
```

```python
result = pd.DataFrame({"단어": dic.keys(),
                       "빈도": dic.values()})
result.sort_values('빈도',ascending=False)
```

```python
from collections import Counter

from collections import Counter

result = Counter(words)
result.most_common(20)
```

```python
from PIL import Image
import numpy as np
img = Image.open('../data/1_gW5CuoIxGQknHNjv7mqRpA.png')
img = np.array(img)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color='white', font_path='../Day_02/BMDOHYEON_ttf.ttf',mask=img)
cloud = wc.generate_from_frequencies(result)
plt.imshow(cloud)
plt.show()
```
