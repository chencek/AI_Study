```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as rc
from matplotlib import font_manager

fe = font_manager.FontEntry(
    fname=r'C:\Users\KDP-26-\AppData\Local\Microsoft\Windows\Fonts\TENADA.ttf', # ttf 파일이 저장되어 있는 경로
    name='TENADA')                        # 이 폰트의 원하는 이름 설정
font_manager.fontManager.ttflist.insert(0, fe)              # Matplotlib에 폰트 추가
plt.rcParams.update({'font.size': 12, 'font.family': 'TENADA'}) # 폰트 설정
# plt.rc('font',family = 'Malgun Gothic')
person_daegu = pd.read_excel('가구원수별 현황.xlsx')
person_daegu = person_daegu.iloc[[7]].set_index('구분')
person_daegu = person_daegu.applymap(lambda x: int(x))
person_daegu.iloc[0].values
plt.plot(person_daegu.columns.values,person_daegu.iloc[0].values,marker='o',color='slategray')
plt.title('년도별 1인가구 수')
plt.xlabel('년도')
plt.ylabel('인구수')
print(person_daegu)
```

```python
people_daegu = pd.read_excel('세대원수별 세대수.xlsx')
people_daegu = people_daegu.iloc[0,:]
total =int(people_daegu.values[1].replace(',',''))
print(total)
```

```python
total_minus_one= total-person_daegu[2021].values[0]
population = [total_minus_one, person_daegu[2021].values[0]]

# plt.rc('font', family = 'Malgun Gothic')
color  =['#BDBDBD', '#DF0101']
plt.pie(population, labels=['전체가구-1인가구','1인가구'], autopct='%.1f%%', colors=color, startangle=90,wedgeprops = {"edgecolor" : "lightslategrey",'linewidth': 1,'antialiased': True},explode=(0,0.1))
plt.legend(loc = 'upper left')
plt.title('전체 가구 대비 1인가구 비율')
plt.show
```
